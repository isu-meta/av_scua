```python
import pandas as pd
df = pd.read_excel('Master_AV_List.xlsx')
```

```python
def replace_x_with_zeros(x):
    if x=='':
        return '00:00:00'
df = df.fillna('')
df['Duration'] = df['Duration'].apply(lambda x: replace_x_with_zeros(x))
```
